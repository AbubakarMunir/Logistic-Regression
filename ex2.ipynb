```python
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv('ex1data2.txt',usecols=[0,1,2],header=None)
df = pd.DataFrame(data)
#number of features
n = 2
i = 0
mean = []
sd = []
while i <n:
    mean.append(df[i].mean())
    sd.append(df[i].std())
    df[i] = df[i]-mean[i]
    df[i] = df[i]/sd[i]
    i += 1





```

```python
#cost calculation function
import numpy as np
#cost calculation for single iteration
height = len(df[0])
width = 1
d = pd.DataFrame(1 ,index=range(height), columns=range(width))
newdf = [d[0],df[0], df[1]].copy()
newdf = pd.concat(newdf,axis=1)
newdf=newdf.values


def calculate_cost(y,features,weights):
    n=len(weights)
    cost=0
    for x,y in zip(features,y):
        y_pred = np.dot(np.array([x]),weights)
        cost=(y_pred - y)**2
    cost = cost / (2*len(df[0]))
    return cost
calculate_cost(df[2],newdf,np.array([1,1,1]))

```
